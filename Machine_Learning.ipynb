2. What is correlation?What does negative correlation mean?


-> Correlation is a statistical measure that describes how two variables move in relation to each other. It tells us if one variable tends to go up or down when another variable changes. It is measured on a scale from -1 to 1. A correlation of 1 means they move perfectly together, while -1 means they move in opposite directions.


-> Negative correlation means that as one variable increases, the other variable decreases. It indicates an inverse relationship. A real-world example would be the relationship between the temperature outside and the sales of hot tea.

4. How does loss value help in determining whether the model is good or not?

-> The loss value represents the penalty for a bad prediction. It is a number indicating how far off the model's predictions are from the actual values. A high loss value means the model is making large errors. A low loss value means the model's predictions are close to the truth.

5. What are continuous and categorical variables?

-> Continuous variables are numeric values that can be measured and can take any value within a range. Ex- height, weight, temperature, or price. We can have 10.5, 10.55, or 10.552, making them continuous.


Categorical variables represent groups or categories. They usually have a limited number of possible values. Ex- gender (Male/Female).

6. How do we handle categorical variables in Machine Learning? What are the common techniques?

-> Machine learning models generally understand numbers, not words. So we must convert categorical text data into numbers. The common techniques are:

Label Encoding: Assigning a unique number to each category (e.g., Red=1, Blue=2). This is good for ordinal data where order matters (like Low, Medium, High).

One-Hot Encoding: Creating a new column for each category with binary values (0 or 1). For example, a "Color" column becomes "Is_Red" and "Is_Blue" columns. This is better for nominal data where order doesn't matter

7. What do you mean by training and testing a dataset?

-> Training dataset: This is the larger portion (usually 70-80%) used to teach the model. The model sees the answers here and learns the patterns.

-> Testing dataset: This is the remaining portion used to evaluate the model. The model has never seen this data before. We use it to simulate how the model will perform in the real world.

15. What is causation? Explain difference between correlation and causation with an example

-> Causation means that one event actually causes the other to happen.

Correlation: Two things happen at the same time or move together.

Causation: Action A directly results in Action B.

Ex: Ice cream sales and shark attacks are highly correlated (both go up in summer). However, eating ice cream does not cause shark attacks. The cause for both is a third factor: hot weather. Correlation does not imply causation

16. What is an Optimizer? What are different types of optimizers? Explain each with an example

-> An optimizer is an algorithm that updates the parameters of the model (like weights) to minimize the loss function. It decides how to change the model to make it more accurate.

-> Gradient Descent: The most basic type. It takes small steps downhill towards the lowest error. Think of walking down a mountain step-by-step.

-> Stochastic Gradient Descent (SGD): Instead of using all data to decide the next step, it uses just one random sample. It is faster but noisier.

-> Adam (Adaptive Moment Estimation): A smart optimizer that adjusts the speed of learning for each parameter individually. It is like walking down the mountain but being able to run on flat parts and slow down on steep, dangerous parts.

21. What is feature scaling? How does it help in Machine Learning?

-> Feature scaling is the process of putting all our input variables on the same scale (usually between 0 and 1 or centered around 0). It helps because many algorithms (like Gradient Descent or KNN) calculate distances between data points. If one feature is "Salary" (ranging 20,000 to 100,000) and another is "Age" (ranging 20 to 60), the Salary will dominate the calculation simply because the numbers are bigger. Scaling ensures all features contribute equally.

25. Explain data encoding?

-> Data encoding is the process of converting categorical data (text or labels) into a numerical format that a computer can understand. Since machine learning models perform mathematical operations, they cannot process strings like "Apple" or "Orange."
