3. What are the different ways to define and call a function in Python?
- def statement: The most common way, used to create a named function.

- lambda keyword: Used to create small, one-line, anonymous functions.
- The different ways to call function are -
  - Positional Arguments: Arguments are passed in the same order as the parameters are defined.

  - Keyword Arguments: Arguments are passed by explicitly naming the parameter, so order do not matter.

  - Default Arguments: The function definition provides a default value for a parameter, making that argument optional when the function is called.

```python
# Example code for Q3
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

multiply = lambda x, y: x * y


# 1. Positional arguments
print(greet("Aditya"))

# 2. Keyword arguments (order do not matter)
print(greet(greeting="Hi", name="Kiran"))

# 3.lambda Function
print(f"5 * 6 = {multiply(5, 6)}")
```

6. Explain the concept of generators in Python and how they are defined.
- Generators are the function that pauses after every execution and resumes,i.e instead of giving the result in one go it iterates and yields the result one by one.
- it involves the use of yield keyword to yield one value at a time instead of using the return keyword



```python
# Ex code for Q6
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

counter = count_up_to(3)
print(f"From function: {next(counter)}")
print(f"From function: {next(counter)}")
```

7.  What are the advantages of using generators over regular functions?
- The main advantages of using generators instead of regular function are-
  - Memory Efficient-> The generators do not involve the use of sequnce of memory locations instead the same memory location is used or alloted as per the demand thus making it suitable for use of large Datasets.
  - Simpler code-> They are often simpler to write than creating a full custom iterator class (which would require defining __iter__() and __next__() methods).

```python
# Ex code for Q7
import sys

def first_n_numbers_list(n):
    nums = []
    for i in range(n):
        nums.append(i)
    return nums

def first_n_numbers_gen(n):
    for i in range(n):
        yield i

n = 1_000_000

list_data = first_n_numbers_list(n)
gen_data = first_n_numbers_gen(n)

print(f"List size: {sys.getsizeof(list_data)} bytes")
print(f"Generator size: {sys.getsizeof(gen_data)} bytes")
```

10. 10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

- map(func, iter): Transforms each element. It applies func to every item and returns an iterator of the results.
- filter(func, iter): Selects elements. It applies func to every item and keeps only those for which func returns True.
- reduce(func, iter): Combines elements. It cumulatively applies func to the items, reducing the entire sequence down to a single value

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x*x, numbers))
print(f"map: {squares}")
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(f"filter: {evens}")
total_sum = reduce(lambda x, y: x + y, numbers)
print(f"reduce: {total_sum}")
```

1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_even_numbers(numbers):
  return sum(num for num in numbers if num % 2 == 0)

my_list = [1, 2, 3, 4, 5, 6, 7, 8]
print(f"Sum of evens: {sum_even_numbers(my_list)}")
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
  if n <= 1:
    return False
  if n == 2:
    return True
  if n % 2 == 0:
    return False

  for i in range(3, int(n**0.5) + 1, 2):
    if n % i == 0:
      return False

  return True

print("Primes from 1 to 50:")
primes = [num for num in range(1, 201) if is_prime(num)]
print(primes)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
def fib_generator(max_terms):
  a, b = 0, 1
  for _ in range(max_terms):
    yield a
    a, b = b, a + b
print("First 8 Fibonacci numbers:")
fib_sequence = fib_generator(8)

for number in fib_sequence:
  print(number, end=" ")
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
  for i in range(max_exponent + 1):
    yield 2**i

for num in powers_of_two(5):
  print(num, end=" ")
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
with open("sample.txt", "w") as f:
  f.write("This is the first line.\n")
  f.write("This is the second line.\n")
  f.write("This is the third line.\n")

def read_lines_gen(filename):
  try:
    with open(filename, 'r') as f:
      for line in f:
        yield line.strip()
  except FileNotFoundError:
    print(f"Error: File '{filename}' not found.")
print(f"\n\nContents of sample.txt:")
for line in read_lines_gen("sample.txt"):
  print(line)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
celsius_temps = [0, 10, 20, 37, 100]
fahrenheit_temps = map(lambda c: (c * 9/5) + 32, celsius_temps)
print(f"\nCelsius: {celsius_temps}")
print(f"Fahrenheit: {list(fahrenheit_temps)}")
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
my_string = "Hello, what a wonderful day!"
vowels = "aeiouAEIOU"

filtered_chars = filter(lambda char: char not in vowels, my_string)

result_string = "".join(filtered_chars)

print(f"\nOriginal string: {my_string}")
print(f"Without vowels: {result_string}")
```
