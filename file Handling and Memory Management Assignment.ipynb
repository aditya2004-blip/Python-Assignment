15.  How does Python's garbage collection system work?
- Python's garbage collection uses the concept of Reference Counting
  - Every object has a reference count
  - The value of Reference count increases with the increase in the number of variables pointing out to the object.
  - It decreases when a variable stops pointing to it
  - When the reference count drops to zero, the object is immediately deallocated.

4.  Write a Python script that reads from one file and writes its content to another file

```python
source_filename = "source.txt"
dest_filename = "destination.txt"

try:
    with open(source_filename, "r") as src:
        with open(dest_filename, "w") as dest:
            dest.write(src.read())
    print("Successfully copied.")
except FileNotFoundError:
    print("The source file was not found.")
```

6. Write a Python program that logs an error message to a log file when a division by zero exception occurs

```python
import logging
logging.basicConfig(filename="error.log",level=logging.ERROR, format="%(asctime)s - %(levelname)s - %(message)s")

try:
    result = 10 / 0
except ZeroDivisionError:
    logging.error("Attempted to divide by zero.")
```

18.  How would you implement a basic logging setup that logs to a file with rotation after 1MB?


```python
import logging
from logging.handlers import RotatingFileHandler
one_mb=1048576

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[
        RotatingFileHandler('rotating.log', maxBytes=one_mb, backupCount=3)
    ]
)

logger = logging.getLogger()

print("Logging setup complete. Check 'rotating.log'.")

[logger.info(f"This is log message number {i}") for i in range(10000)]
```

21. Write a Python program that reads a file and prints the number of occurrences of a specific word

```python
def count_word(filename, word):
    try:
        with open(filename, 'r') as f:
            return f.read().lower().count(word.lower())
    except FileNotFoundError:
        return 0


file_name = 'sample_text.txt'
word_to_find = 'Python'

count = count_word(file_name, word_to_find)
print(f"'{word_to_find}' count: {count}")
```

22.  How can you check if a file is empty before attempting to read its contents?

```python
import os

filename = "myfile.txt"

try:
    if os.path.getsize(filename) > 0:
        with open(filename, "r") as f:
            print(f.read())
    else:
        print("The file is empty.")
except FileNotFoundError:
    print(f" The file '{filename}' does not exist.")
```
