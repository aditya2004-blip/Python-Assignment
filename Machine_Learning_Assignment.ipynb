1. What is Simple Linear Regression?

-> Simple Linear Regression is a statistical method that allows us to summarize and study relationships between two continuous variables. It attempts to model the relationship between two variables by fitting a straight line to observed data.

2. What are the key assumptions of Simple Linear Regression?

   -> For Linear Regression to be reliable there are 4 key assumptions:-

   a. Linearity: The relationship between $X$ and $Y$ is linear

   b. Independence: The observations are independent of each other

   c. Homoscedasticity: The variance of the errors is constant across all levels of $X$.

   d. Normality: The residuals of the model are normally distributed.


3.  What does the coefficient m represent in the equation Y=mX+c?

  -> The coefficient $m$ represents the slope of the regression line. It tells us how much the dependent variable ($Y$) changes for every single unit increase in the independent variable ($X$).

4. What does the intercept c represent in the equation Y=mX+c?

-> The intercept $c$ represents the expected value of $Y$ when $X$ is exactly 0. It is the point where the regression line crosses the Y-axis.

5.  How do we calculate the slope m in Simple Linear Regression?

-> we calculate the slope $m$ by dividing the covariance of $X$ and $Y$ by the variance of $X$.$$m = \frac{\sum((X - \text{mean}(X)) \times (Y - \text{mean}(Y)))}{\sum((X - \text{mean}(X))^2)}$$

6. What is the purpose of the least squares method in Simple Linear Regression?

-> The "Least Squares" method is the technique we use to find the best-fitting line. It works by minimizing the sum of the squares of the vertical differences between the observed data points and the fitted line.

7.  How is the coefficient of determination (R²) interpreted in Simple Linear Regression?

->  $R^2$ tells us the proportion of the variance in the dependent variable ($Y$) that is predictable from the independent variable ($X$). It ranges from 0 to 1.

8. What is Multiple Linear Regression?

-> Multiple Linear Regression is an extension of simple linear regression. Instead of using just one independent variable to predict $Y$, we use two or more independent variables ($X_1, X_2, \dots$). The goal is to model the relationship between a single continuous dependent variable and multiple independent variables.

9.  What is the main difference between Simple and Multiple Linear Regression?

-> The main difference lies in the number of inputs. Simple Linear Regression uses only one feature to make predictions.Multiple Linear Regression uses multiple features (fitting a plane or hyperplane in multi-dimensional space).

10.  What are the key assumptions of Multiple Linear Regression?

->  We rely on the same assumptions as simple linear regression (Linearity, Independence, Homoscedasticity, Normality) plus one new important assumption:

-> No Multicollinearity: The independent variables should not be highly correlated with each other.

11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?

-> Heteroscedasticity happens when the noise or error in our model varies depending on the value of the input. For example predicting house prices might be very accurate for small houses but very wild for mansions.If present our coefficient estimates might still be unbiased, but our standard errors will be wrong, making hypothesis tests unreliable.

12.  How can you improve a Multiple Linear Regression model with high multicollinearity?

-> Remove one of the highly correlated variables.

-> Combine the correlated variables into a single feature.

-> Use dimensionality reduction techniques like Principal Component Analysis (PCA).

-> Use regularization techniques like Ridge or Lasso regression.

13.  What are some common techniques for transforming categorical variables for use in regression models?

-> The common techniques include:

-> One-Hot Encoding: Creating binary columns (0 or 1) for each category (ex, "Color_Red", "Color_Blue").

-> Label Encoding: Assigning a number to each category (ex,Low=0,Med=1, High=2), which is useful if the categories have a natural order.

14. What is the role of interaction terms in Multiple Linear Regression?

-> Sometimes, the effect of one variable on $Y$ depends on the value of another variable. For Ex-, Ad Spend might increase sales more effectively if Season is Summer. An interaction term (created by multiplying the two variables) allows the model to capture this combined effect.

15. How can the interpretation of intercept differ between Simple and Multiple Linear Regression?

-> In Simple Regression, the intercept is the value of $Y$ when the single $X$ is 0. In Multiple Regression, the intercept represents the expected value of $Y$ when all the independent variables ($X_1, X_2, \dots$) are simultaneously 0.

16. What is the significance of the slope in regression analysis, and how does it affect predictions?

-> The slope indicates the importance of a feature. A steep slope (large number) means a small change in that feature will cause a large change in the prediction and If the slope is near zero, that feature has very little effect on the outcome.

17. How does the intercept in a regression model provide context for the relationship between variables?

-> The intercept anchors the regression line.It provides the baseline prediction if we had no input data. For instance, in a salary prediction model, the intercept might represent the "base salary" before adding bonuses for experience or education.

18. What are the limitations of using R² as a sole measure of model performance?

->  $R^2$ always increases (or stays the same) as we add more variables, even if those variables are useless junk. This can trick us into thinking a complex model is better. It also doesn't tell us if our model is biased or if the predictions are actually close to the real values just how much variance is explained.

19. How would you interpret a large standard error for a regression coefficient?

->  A large standard error means our estimate of that coefficient is not very precise. It suggests that if we took a different sample of data the calculated slope could change drastically. Usually this leads to a high p-value meaning the variable might not be statistically significant.

20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?

-> We identify it by plotting residuals against predicted values. If we see a "funnel shape" (errors get wider as predictions get larger) that is heteroscedasticity. We must address it because it violates the assumption of constant variance meaning our confidence intervals and hypothesis tests might be wrong.

21. What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?

-> This is a classic sign of overfitting with too many unnecessary variables. The high normal $R^2$ is inflated because we added many variables, but the Adjusted $R^2$ penalizes us for adding variables that don't actually help prediction, pulling the score down.

22.  Why is it important to scale variables in Multiple Linear Regression?

->  Scaling (normalization/standardization) is crucial when variables have different units. Without scaling, the coefficient for "Salary" might look tiny compared to "Age" just because the numbers are larger, making it hard to compare their importance. It also helps optimization algorithms (like Gradient Descent) converge faster.

23.  What is polynomial regression?

->  Polynomial regression is a form of regression analysis where the relationship between the independent variable $X$ and the dependent variable $Y$ is modeled as an $n$-th degree polynomial. Instead of a straight line, it fits a curve to the data.

24. How does polynomial regression differ from linear regression?

->  Linear regression finds the best straight line to fit the data. Polynomial regression finds the best curved line. It allows for more complex, non-linear relationships.

25.  When is polynomial regression used?

->  We use it when the data points clearly do not follow a straight line. For example, the growth of bacteria over time or the trajectory of a ball thrown in the air (which follows a curve) are better modeled with polynomial regression.

26. What is the general equation for polynomial regression?

-> The equation looks like this:$$Y = \beta_0 + \beta_1X + \beta_2X^2 + \dots + \beta_nX^n + \epsilon$$Here, we are still doing linear regression, but we create new features by squaring or cubing the original $X$.

27.  Can polynomial regression be applied to multiple variables?

-> Yes. We can have polynomial terms for multiple variables (ex,$X_1^2$, $X_2^2$) and even interaction terms between them. However, this increases the complexity of the model very quickly.

28. What are the limitations of polynomial regression?

-> Overfitting: High-degree polynomials can become very "wiggly" and try to pass through every single data point, capturing noise instead of the pattern.

-> Extrapolation: Polynomial models are terrible at predicting values outside the range of the data used to train them. The curve will shoot off to infinity very quickly.

29. What methods can be used to evaluate model fit when selecting the degree of a polynomial?

->  We typically use Cross-Validation. We test different degrees (linear, quadratic, cubic, etc.) and see which one performs best on a testing dataset. We look for the sweet spot where the error is low on both training and testing data avoiding the high variance of higher-degree polynomials.

30. Why is visualization important in polynomial regression?

->  Since polynomial models can take on strange shapes, visualization helps us verify that the curve makes sense. It allows us to see if the model is capturing the underlying trend or just connecting the dots (overfitting) in a way that doesn't represent reality.

31. How is polynomial regression implemented in Python?


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1, 4, 9, 16, 25])  # y = x^2

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

y_pred = model.predict(X_poly)

print("Predictions:", y_pred)
```
